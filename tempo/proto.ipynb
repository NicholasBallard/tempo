```python
import inspect
from tempo import RateLimiter
from functools import wraps
from functools import partial
from typing import Awaitable, Callable, Iterable
import time
import httpx
from pydantic import BaseModel, validator, AnyHttpUrl
import asyncio
import nest_asyncio
nest_asyncio.apply()


uri = 'https://jsonplaceholder.typicode.com/todos/'


async def afetch(index: int):
    async with httpx.AsyncClient() as client:
        await asyncio.sleep(1)
        res = await client.get(uri + f'{index}')
        return res


# begin - collector function
def add_to_list(list_: list, obj: object) -> None:
    list_.append(obj)


responses = []
list_add = partial(add_to_list, responses)
# end - collector function


# iterable
rng = range(1, 11)


# wait for all the tasks to complete
async def main(function: Callable, iterable: Iterable, *iterables: Iterable):
    tasks: Iterable[Awaitable] = map(function, iterable, *iterables)
    responses = await asyncio.gather(*tasks)
    return responses


# greedily fetch the coroutines as they complete
async def main(function: Callable, iterable: Iterable, collector: Callable, *iterables: Iterable):
    tasks: Iterable[Awaitable] = map(function, iterable, *iterables)
    for coroutine in asyncio.as_completed(tasks):
        res = await coroutine
        collector(res)


def res():
    start_time = time.monotonic()
    res = asyncio.run(
        main(afetch, rng, list_add)
    )
    print(f'Elapsed: {time.monotonic() - start_time}')
```
